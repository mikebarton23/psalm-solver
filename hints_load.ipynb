```python
import pandas as pd
```

```python
df = pd.read_json("data/hints.json", orient='index', dtype=str)
```

```python
# Convert the index from strings to integers
df.index = df.index.astype(int)
```

```python
def clean_index(index):
    # Convert to string, remove the trailing '.0', and convert back to integer
    return int(str(index).split('.')[0])
```

```python
# Apply the function to each index
df.index = df.index.map(clean_index)
```

```python
# Convert the index to strings, slice the first 8 characters, and convert back to integers
df.index = df.index.astype(str).str.slice(0, 8).astype(int)
```
